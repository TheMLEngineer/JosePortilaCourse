```python
f = open('Contact_Email_Information.pdf' , 'rb')
```

```python
pdf_reader = PyPDF2.PdfFileReader(f)
```

```python
pdf_reader
```

```python
pdf_reader.numPages
```

```python
pdf_reader.getPage(1).extractText()
```

```python
pdf_text = []
for n in range(pdf_reader.numPages):
    
    page = pdf_reader.getPage(n)
    pdf_text.append(page.extractText())
```

```python
pdf_text
```

```python
import re
```

```python
for list_element in pdf_text:
        print(re.findall(r'[\w_.-]+@[\w_.-]+' , list_element))
```
